# 1. Load Dataset

```python
import pandas as pd
df = pd.read_csv(r"C:\Users\nisar\Downloads\student-mat.csv",sep=";")
df.head()
```

# 2. Explore & Clean Data

```python

# Check dataset shape
print("Shape:", df.shape)
df.dtypes

# Check missing values
df.isnull().sum()

# Remove duplicate rows
df = df.drop_duplicates()

print("After removing duplicates:", df.shape)
```

# 3. Average Final Grade (G3)

```python

avg_g3 = df["G3"].mean()

avg_g3
```

```python
# Number of Students Scoring Above 15

above_15 = df[df["G3"] > 15].shape[0]
above_15
```

```python
# Correlation Between Study Time and Final Grade

correlation = df["studytime"].corr(df["G3"])
correlation
```

```python
# Average Final Grade by Gender

gender_avg = df.groupby("sex")["G3"].mean()
gender_avg
```

# 4. Visualizations

```python

import matplotlib.pyplot as plt

```

```python
# Histogram of Final Grades

plt.figure()
plt.hist(df["G3"], bins=10)
plt.xlabel("Final Grade (G3)")
plt.ylabel("Number of Students")
plt.title("Distribution of Final Grades")
plt.show()
```

```python
# Scatterplot: Study Time vs Grades

plt.figure()
plt.scatter(df["studytime"], df["G3"])
plt.xlabel("Study Time")
plt.ylabel("Final Grade (G3)")
plt.title("Study Time vs Final Grade")
plt.show()
```

```python
# Bar Chart: Male vs Female Average Score

plt.figure()
gender_avg.plot(kind="bar")
plt.xlabel("Gender")
plt.ylabel("Average Final Grade")
plt.title("Average Final Grade by Gender")
plt.show()
```

# Conclusion

[1] The student performance data is successfully loaded, cleaned, and visualized using pandas.

[2] The average final grade (G3) gave an indication of the general level of academic achievement by students.

[3] A number of students had grades above 15. This shows levels of good academic performance.

[4] Study time was observed to be positively correlated to grades obtained, implying that increased study time translates to higher performance.

[5] There was a variation in the mean performance of male and female students.

[6] Graphs and other visualization techniques were very helpful in making sense of the grade distributions and important relationships.
